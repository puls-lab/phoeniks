# Example 5 of Phoeniks
> **P**ULS **h**ands-on **o**ptimized **e**xtraction *of* $\boldsymbol{n} - \boldsymbol{i}\cdot \boldsymbol{k}$'s


Extract refractive index and absorption from multiple, real-world measurements. We will use these multiple traces to generate the standard deviation for each point in time and thus, the stability of the THz trace at each frequency point. This allows us to make a more informed smoothing when the phase is fluctuating, called the spatial variant, moving average filter `SVMAF`.

```python
# General Python libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import EngFormatter
%matplotlib notebook

home_dir = os.getcwd()
os.chdir(r"../..")
# Own library
import phoeniks as pk
os.chdir(home_dir)
```

```python
help(pk)
```

# Load data
Loading reference file.

```python
ref_file = "td_reference_mean.txt"
ref = np.loadtxt(ref_file)
ref, ref.shape
```

The first column contains the signal in [s], the second column the signal in [V].
We can use the first column to derive the time-axis for the signal and save it seperately:

```python
time = 1e-12 * ref[:, 0]
```

Load the sample file. For this real measurement of the organic crystal BNA, the thickness is 450 µm.

```python
sample_thickness = 450e-6
sam_file = "td_sample_mean.txt"
sam = np.loadtxt(sam_file)
sam, sam.shape
```

```python
dark_file = "td_dark_mean.txt"
dark = np.loadtxt(dark_file)
dark, dark.shape
```

```python
fd_sample_std_real, fd_sample_std_imag = np.loadtxt("fd_sample_std.txt")
fd_reference_std_real, fd_reference_std_imag = np.loadtxt("fd_reference_std.txt")
fd_dark_std_real, fd_dark_std_imag = np.loadtxt("fd_dark_std.txt")
```

# Create data object
Feed this data into the THz data object. The following data needs is supplied:
1. time-axis (same for reference and sample trace)
2. Signal of reference measurement in time-domain
3. Signal of sample measurement in time-domain
4. Dark signal (blocked THz beam) in time-domain
5. Standard deviation (real and imaginary) of reference measurement in frequency-domain
6. Standard deviation (real and imaginary) of sample measurement in frequency-domain
7. Standard deviation (real and imaginary) of dark measurement in frequency-domain

```python
data = pk.thz_data.Data(time = time, 
                        td_reference = ref[:, 1], 
                        td_sample = sam[:, 1],
                        td_dark = dark[:, 1],
                        fd_reference_std = fd_reference_std_real + 1j * fd_reference_std_imag,
                        fd_sample_std = fd_sample_std_real + 1j * fd_sample_std_imag,
                        fd_dark_std = fd_dark_std_real + 1j * fd_dark_std_imag)
```

```python
data.window_traces(time_start=-740e-12, time_end=-695e-12)
data.pad_zeros(new_frequency_resolution=2e9)
```

# Plot data
To get a better view on the data, we can create a plot of reference and sample file:

```python
plot = pk.plot.Plot()
plot.plot_data(data)
```

## Extraction of optimal thickness

```python
extract_obj = pk.extraction.Extraction(data)
```

```python
extract_obj.unwrap_phase(frequency_start=0.6e12, frequency_stop=2e12)
```

```python
plot.plot_phase(extract_obj)
```

```python
print(f"Sample thickness: {EngFormatter('m')(sample_thickness)}")
```

```python
n, k = extract_obj.get_initial_nk(thickness=sample_thickness)
delta_max = extract_obj.get_max_delta(thickness=sample_thickness)
```

```python
thickness_array, tv_dict = extract_obj.get_thickness_array(thickness=sample_thickness,
                                                           thickness_range=100e-6,
                                                           step_size=5e-6)
```

```python
fig, ax = plt.subplots()
ax.plot(thickness_array, tv_dict["tv_1"], label="TotalVariation, deg=1")
ax.plot(thickness_array, tv_dict["tv_2"],label="TotalVariation, deg=2")
ax.plot(thickness_array, tv_dict["tv_n"],label="TotalVariation, Nick's method")
ax.legend()
ax.grid(True)
ax.xaxis.set_major_formatter(EngFormatter("m"))
```

# Extraction
Start the extraction of the refractive index and absorption.
It is necessary, to create an extraction object:

```python
data = pk.thz_data.Data(time = time, 
                        td_reference = ref[:, 1], 
                        td_sample = sam[:, 1],
                        td_dark = dark[:, 1],
                        fd_reference_std = fd_reference_std_real + 1j * fd_reference_std_imag,
                        fd_sample_std = fd_sample_std_real + 1j * fd_sample_std_imag,
                        fd_dark_std = fd_dark_std_real + 1j * fd_dark_std_imag)
data.window_traces(time_start=-740e-12, time_end=-695e-12)
data.offset_time_to_reference_peak()
data.pad_zeros(new_frequency_resolution=2e9)
```

```python
plot = pk.plot.Plot()
plot.plot_data(data)
```

```python
extract_obj = pk.extraction.Extraction(data)
extract_obj.unwrap_phase(frequency_start=0.3e12, frequency_stop=5.7e12)
```

```python
plot.plot_phase(extract_obj)
```

Getting a first guess of n and k, which is necessary for the optimization algorithm:

```python
n, k = extract_obj.get_initial_nk(thickness=sample_thickness)
plot.plot_refractive_index(extract_obj.data.frequency, extract_obj.data.n, extract_obj.data.k)
```

Looking at the time-domain, we cannot only measure the time delta between the peak of reference and sample trace, but also how many echoes (Fabry-Perot reflections) would fit in the remaining time trace:

```python
delta_max = extract_obj.get_max_delta(thickness=sample_thickness)
print(f"Δt:\t\t{EngFormatter('s', places=1)(extract_obj.get_time_shift())}")
print(f"Max. echoes:\t{delta_max}\n")
```

# Run optimization

```python
frequency, n_opt, k_opt, alpha_opt = extract_obj.run_optimization(thickness=sample_thickness)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True)
ax[0].set_title("Optimization of refractive index and absorption")
ax[0].plot(frequency, n_opt, label=r"$n(f)$")
ax[0].xaxis.set_major_formatter(EngFormatter("Hz"))
ax[0].set_xlabel("Frequency")
ax[0].set_ylabel("Refractive Index")
ax[0].grid(True)
ax[0].legend()


ax[1].plot(frequency, 0.01 * alpha_opt, label=r"$\alpha(f)$")
ax[1].xaxis.set_major_formatter(EngFormatter("Hz"))
ax[1].set_xlabel("Frequency")
ax[1].set_ylabel(r"Absorption [$\mathrm{cm}^{-1}$]")
ax[1].grid(True)
ax[1].legend()
plt.tight_layout()
```

```python
svmaf_obj = pk.SVMAF(extract_obj)
svmaf_obj.setup()
n_smooth, k_smooth, alpha_smooth = svmaf_obj.run(thickness = sample_thickness)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True)
ax[0].set_title("Optimization of refractive index and absorption")
ax[0].plot(frequency, n_opt, label=r"$n(f)$")
ax[0].plot(extract_obj.data.frequency, n_smooth, label=r"SVMAF of $n(f)$")
ax[0].xaxis.set_major_formatter(EngFormatter("Hz"))
ax[0].set_xlabel("Frequency")
ax[0].set_ylabel("Refractive Index")
ax[0].grid(True)
ax[0].legend()

ax[1].plot(frequency, 0.01 * alpha_opt, label=r"$\alpha(f)$")
ax[1].plot(extract_obj.data.frequency, 0.01 * alpha_smooth, label=r"SVMAF of $\alpha(f)$")
ax[1].xaxis.set_major_formatter(EngFormatter("Hz"))
ax[1].set_xlabel("Frequency")
ax[1].set_ylabel(r"Absorption [$\mathrm{cm}^{-1}$]")
ax[1].grid(True)
ax[1].legend()
plt.tight_layout()
```

```python
df = pd.DataFrame(data={"frequency" : extract_obj.data.frequency,
                        "n" : n_smooth,
                        "alpha" : 0.01 * alpha_smooth})
df.to_csv("output.csv", header=True, index=False)
```
